# K-Means Clustering

K-Means Clustering is an unsupervised machine learning algorithm used to group similar data points into K clusters.

K = number of clusters you want.

Data points in the same cluster are more similar to each other than to other clusters.

***How it Works (Steps)***

1. Choose K clusters

2. Initialize K centroids (random points)

3. Assign each data point to the nearest centroid

4. Update centroid by calculating the mean of points in that cluster

5. Repeat steps 3–4 until centroids stop changing

👍 Pros

Simple and fast algorithm

Easy to implement and understand

Works well with large datasets

Efficient when clusters are clearly separated

👎 Cons

Must choose K (number of clusters) beforehand

Sensitive to outliers (extreme values affect centroids)

Works best with spherical / evenly sized clusters

Different initial centroids can give different results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Creating the Dataset

A sample dataset is created containing students' study hours and attendance. This dataset will be used to demonstrate how K-Means groups similar observations into clusters.

```python
# Create dataset
data={
    'Study_hours': [1,2,3,4,5,6,7,7,6,1],
    'Attendance': [40,50,60,70,80,90,45,55,65,75]
}
```

# Creating a DataFrame

The dataset is converted into a Pandas DataFrame to make it easier to visualize, manipulate, and apply machine learning algorithms.

```python
# Convert dictionary into pandas dataframe (table format)
df=pd.DataFrame(data)
print("Dataset:")
print(df)
```

# Visualizing the Dataset

Before applying the K-Means algorithm, the data is visualized using a scatter plot to understand its distribution and identify any visible patterns.

```python
#2 Visualize the data
# Create a scatter plot  to visualize the relationship between study hours and attendance
plt.scatter(df['Study_hours'], df['Attendance'])
```

```python
#title of the graph
plt.title('Student Data Distribution')
#Display the graph
plt.show()
#label for X-axis
plt.xlabel('Study Hours')
#label for Y-axis
plt.ylabel('Attendance')
```

# Applying the K-Means Algorithm

The K-Means model is created by specifying the desired number of clusters (`n_clusters`). The algorithm groups similar observations based on their distance from the cluster centroids.

```python
#3 Apply K-means Algorithm
#n_clusters = no. of groups we want to divide the data into
kmeans = KMeans(n_clusters=2)

# Fit the model on the dataset and predict which cluster each data point be
df['Cluster'] = kmeans.fit_predict(df)

# print  dataset again to see which clust each student belongs too
print("\nDataset with Clusters label:")
print(df)
```

# Visualizing the Clusters

After training the model, each data point is assigned to a cluster. The scatter plot is updated to display different colors for different clusters, making the grouping easier to understand.

```python
#4.Visualize clusters
# Create scatter plot again but now color point baseed on cluster no.
plt.scatter(df['Study_hours'], df['Attendance'], c=df['Cluster'])
plt.xlabel('Study Hours')
plt.ylabel('Attendance')
plt.title('K-means Clustering result')
plt.show()
```

# Cluster Centers

Cluster centers (centroids) represent the average position of all data points within each cluster. They serve as the reference points used by the K-Means algorithm during clustering.

```python
#5. Display cluster centers
#print the center point of cluster
# these are the average values of each cluster group
print("\nCluster Centers:")
print(kmeans.cluster_centers_)
#
```
