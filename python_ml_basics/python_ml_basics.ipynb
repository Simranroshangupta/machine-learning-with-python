## Lists

Lists are ordered, mutable collections in Python that can store elements of different data types. They support indexing, slicing, insertion, deletion, and modification.

```python
a=[1,2,3,"simmi",(1,2)]
del a[-1]
print(a)
a[3]=4
print(a)
```

## Tuples

Tuples are ordered and immutable collections. Once created, their elements cannot be modified. They are useful for storing fixed collections of data.

```python
a[1:]=["two","three","four"]
print(a)
```

## Sets

Sets are unordered collections of unique elements. They are useful for removing duplicates and performing mathematical set operations such as union and intersection.

```python
s={1,2,3,4,5,4,3,2,6}
s
```

```python
p={4,5,6,7,8,9,}
```

```python
s.union(p)
```

```python
p.union(s)
```

```python
s.intersection(p)
```

```python
s.difference(p)
```

```python
s|p
```

```python
s&p
```

```python
s-p
```
