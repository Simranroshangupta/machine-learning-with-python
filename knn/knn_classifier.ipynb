# K-Nearest Neighbor(KNN)
K-Nearest Neighbors (KNN) is a classification technique by which we are able to classify a particular new observation into a specific category. KNN can follow **distance matric**.
It is a very **lazy algorithm**

**How KNN works?**

1. choose k(neighbors) and a distance metric

2. calculate distance to every training point

3. pick the k smallest distance

4. vote by majority(classificaton)

5. (for regression: average the values)


**Euclideans distance**:

d = √((x2 − x1)² + (y2 − y1)²)

# Splitting the Dataset

The dataset is divided into two parts:

- **Training Set** – Used to train the machine learning model.
- **Testing Set** – Used to evaluate the model's performance on unseen data.

This helps measure how well the model generalizes to new data.

```python
from sklearn.model_selection import train_test_split  #library → module → function
                                                      #sklearn → model_selection → train_test_split
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

# Building the KNN Model

The K-Nearest Neighbors classifier is created by specifying the number of nearest neighbors (`k`). The model is then trained using the training dataset.

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
```

# Model Training

The training dataset is used to teach the KNN algorithm how to classify observations based on their nearest neighbors.

```python
model.fit(x_train,y_train)
```

# Model Prediction

After training, the model predicts the class labels for the testing dataset.

```python
ypred=model.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,ypred)
```

# Model Evaluation

The performance of the classifier is evaluated using the Accuracy Score, which measures the percentage of correctly classified observations.

```python
acc  #accuracy
```

# Predicting a New Observation

Once the model has been trained and evaluated, it can predict the class of new unseen data based on the learned patterns.

```python
new=[[4,7]]
result=model.predict(new)
print(result)
```
