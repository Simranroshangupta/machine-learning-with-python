# 2. Series

A **Series** is a one-dimensional labeled array capable of storing different types of data.

It consists of:

- Values
- Index

Series are commonly used when working with a single column of data.

```python
import numpy as np
import pandas as pd
```

```python
s = pd.Series() #when there is circular open close bracket it is called as function
```

```python
s
```

```python
print(s)
```

```python
s = pd.Series(data=[1,2,3],index=["a","b","c"])
```

```python
s
```

```python
s = pd.Series(data=[1,2,3],index=["a","b","c","d"])    #Value error because Length of values (3) does not match length of index (4)
```

```python
s = pd.Series(data=[1,2, ,3],index=["a","b","c"])  # Syntax error because you have to add the value manually
```

```python
d={"a":100,"b":200,"c":300}   # Using Dictionaries
s= pd.Series(data=d)
```

```python
s
```

```python
d={"a":100,"b":200,"c":300}   # Using Dictionaries
s= pd.Series(data=d,index=["a","b","c","d"]) # d has no value
```

```python
s
```

```python
d={"a":100,"b":200,"c":300}   # Using Dictionaries
s= pd.Series(data=d,index=["a","b","d"]) # excluded c key
```

# 3. DataFrame

A **DataFrame** is a two-dimensional tabular data structure consisting of rows and columns.

It is the most commonly used object in Pandas and is similar to a spreadsheet or SQL table.

# DataFrame   
Data frame is as 2D structure data , data is aligned in a table format which is rows and columns

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
print(df)
```

```python
d= [1,2,3,4,5]
df= pd.DataFrame(d)
df
```

```python
dt = [["i","ii","iii"],["alpha","beta","gamma"],["a","b","c"]]  #Multicolumn execution is possible
df = pd.DataFrame(dt,columns=["col1","col2","col3"])
df
```

```python
a = ["apple","banana","cherry","leechi","grapes","fatiiiimuuu"]
b = [10,20,30,40,50,60]
c = [True,False,True,False,True,False]

dic = {"fruits_name":a,"numbers":b,"boolean":c}
df = pd.DataFrame(dic)
```

```python
df
```

```python
df[df.fruits_name=="leechi"]
```

```python
df[df.numbers<=40]
```

```python
df[df.boolean==True]
```

```python
df[df.fruits_name=="plum"] # no plum in the table
```

```python
df[df.fruits_name=="fatiiiimuuu"]
```

```python
df= pd.DataFrame(np.arange(1,5).reshape(2,2),columns=["a","b"]) # Dataframe creation using numpy
df
```

```python
d={
    "one":pd.Series([1,2,3],index=["a","b","c"]),
    "two":pd.Series([4,3,2,1],index=["a","b","c","d"])
}
df=pd.DataFrame(d)
df
```
