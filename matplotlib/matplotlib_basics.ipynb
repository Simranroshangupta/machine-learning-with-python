# **Line plot**
A line plot is used to visualize trends or changes in data over a continuous interval. It is commonly used to represent time-series data or compare values over a sequence.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,7,5,3,8,2]
```

```python
plt.plot(x)
plt.show()
```

```python
a= np.arange(0,10,0.1)
b= np.sin(a)
```

```python
plt.plot(a,b)
```

```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]
```

```python
plt.plot(area,price)
plt.title("Bsc AIML")
plt.xlabel("Area in Sqrt")
plt.ylabel("Price in lakhs ")

plt.xlim((400,2000))  #limit for x axis
plt.ylim((10,50))

plt.xticks(rotation=45)
plt.yticks(rotation=50)

plt.grid(True)
```

# Multiple Plots (Subplots)

Multiple plots allow us to display two or more graphs within the same figure. This is useful for comparing datasets, visualizing different relationships, or presenting multiple charts together.

Matplotlib provides the `subplot()` and `subplots()` functions to organize multiple plots in a single figure.

### Applications

- Comparing multiple datasets
- Visualizing different graph types together
- Building dashboards
- Exploratory Data Analysis (EDA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= np.arange(1,20)
```

```python
x
```

```python
y1= np.sqrt(x)
y1
```

```python
y2=np.log(x)
y2
```

```python
plt.plot(x,y1,color="blue")
plt.plot(x,y2,color="red")
plt.legend(labels=["Square root","Log"])
```

```python
x=[1,2,3]
y=[5,7,4]

x1=[1,2,3]
y1=[11,14,12]

plt.plot(x,y,label="Line 1",color="blue")
plt.plot(x1,y1,label="Line 2",color="red")
plt.legend()
plt.title("BSC AI/ML")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

# Bar Chart

A bar chart represents categorical data using rectangular bars. The height or length of each bar corresponds to the value it represents.


```python
a, b=[2,4,6,8,10], [10,20,30,40,5]
```

```python
plt.bar(a,b,label="bar chart",color="indigo")
plt.title("Sales Report")
plt.xlabel("Team Numbers")
plt.ylabel("Total Sales")
plt.legend()
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]

x1=[11,16,13,19]
y1=[15,13,17,11]

plt.bar(x,y,label="Line 1",color="magenta")
plt.bar(x1,y1,label="Line 2",color="black")
plt.title("Ai/Ml")
plt.legend()
```

```python
age=[21,23,24,25,26,21,27,29,21,18,14,27,29,29,27]

ids=[x for x in range(len(age))]
```

```python
ids
```

```python
plt.bar(ids,age,color="orange")
plt.xlim()
```

# Histogram

A histogram displays the frequency distribution of numerical data by grouping values into intervals called bins.

```python
age=[21,23,24,25,26,21,27,29,21,18,14,27,29,29,27]
plt.hist(age)
```
