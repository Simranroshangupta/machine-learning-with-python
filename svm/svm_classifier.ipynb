# Support Vector Machine (SVM) Classification

## Overview

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It identifies the optimal decision boundary (hyperplane) that best separates data points belonging to different classes.

---

## Objectives

- Understand the Support Vector Machine (SVM) algorithm
- Load and inspect a dataset
- Separate features and target variables
- Split the dataset into training and testing sets
- Build an SVM classifier
- Train the model
- Make predictions
- Evaluate model accuracy
- Predict the class of a new observation

---

## Dataset

**student_svm_dataset.csv**

### Features

- Study_Hours
- Attendance

### Target

- Result

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC #Support Vector Classifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

import warnings
warnings.filterwarnings("ignore")
```

# Loading the Dataset

The dataset is loaded using Pandas. It contains students' study hours and attendance records, which are used to predict whether a student will pass or fail.

```python
#Load dataset
data = pd.read_csv("/content/student_svm_dataset.csv")
```

# Feature Selection

Machine learning models require:

- **Features (X)** – Input variables used for prediction.
- **Target (y)** – Output variable that the model learns to predict.

In this example:

- Features: `Study_Hours`, `Attendance`
- Target: `Result`

```python
#Seperate features from data
x = data[['Study_Hours','Attendance']]
y = data['Result']
data
```

# Splitting the Dataset

The dataset is divided into two subsets:

- **Training Set** – Used to train the SVM model.
- **Testing Set** – Used to evaluate the model's performance on unseen data.

```python
#Split dataset into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=45,test_size=0.2)
```

# Building the SVM Model

An SVM classifier is created using a **linear kernel**. The model learns the optimal decision boundary that separates different classes in the dataset.

```python
#Create SVM model
model = SVC(kernel='linear')
```

# Model Training

The SVM model is trained using the training dataset. During training, the algorithm identifies the best hyperplane to classify the data.

```python
#train model
model.fit(x_train,y_train)
```

# Model Prediction

After training, the SVM model predicts the class labels for the testing dataset.

```python
# Predict result
predictions = model.predict(x_test)
```

# Model Evaluation

The performance of the classifier is measured using **Accuracy Score**, which indicates the percentage of correctly classified observations.

```python
#Check Accuracy
accuracy = accuracy_score(y_test,predictions)

print("Predictions:", predictions)
print("Accuracy:", accuracy)
```

# Predicting a New Observation

Once trained, the SVM model can classify new unseen observations based on the learned decision boundary.

```python
#Predict new student
new_student = [[4,60]]
result = model.predict(new_student)
print("Prediction for new student:", result)
```
