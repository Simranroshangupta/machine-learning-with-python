# Importing Libraries

The required Python libraries are imported for data manipulation, visualization, preprocessing, model building, and evaluation.

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings("ignore")
```

# Loading the Dataset

The Titanic dataset is loaded into a Pandas DataFrame to inspect its structure and prepare it for preprocessing.

```python
#Load Dataset
data = pd.read_csv('/content/titanic (2).csv')
data
```

```python
data.shape
```

# Feature Engineering

A new feature named **FamilySize** is created by combining the number of siblings/spouses and parents/children travelling with each passenger.

Feature engineering helps improve model performance by creating meaningful information from existing data.

```python
# Create new feature: FamilySize
# FamilySize = siblings/spouse + parents/children + passenger
data["FamilySize"] = data["SibSp"] + data["Parch"] + 1
```

# Feature Selection

Relevant numerical features are selected for model training. The categorical feature **Sex** is also retained because it has a significant influence on survival prediction.

```python
# keep only numeric columns and also Add Sex because it strongly affects survival.
#Because Women survived more and Men died more
data = data[["Pclass","Sex","Age","FamilySize","Fare","Survived"]]
```

# Handling Missing Values

Missing values in the **Age** column are replaced to ensure that the dataset is complete before training the machine learning model.

```python
# fill missing age
data["Age"] = data["Age"].fillna(data["Age"].mean())
#data["Age"].mean() → calculates the average age of all passengers
#fillna() → replaces all missing Age values with 29.7
```

# Encoding Categorical Data

Machine learning algorithms require numerical input. Therefore, categorical values such as **Male** and **Female** are converted into numerical representations.

```python
# convert male/female to numbers
data["Sex"] = data["Sex"].map({"male":0,"female":1})
data
```

# Defining Features and Target

The dataset is divided into:

- **Features (X)** – Input variables used for prediction.
- **Target (y)** – Survival status of passengers.

```python
# Remove the "Survived" column from the dataset to create input features (X)
# "drop" means remove a column from the table
# X contains passenger information used to make predictions
# Y contains the result we want to predict (0 = not survived, 1 = survived)

x = data.drop("Survived", axis=1)
y = data["Survived"]
```

# Splitting the Dataset

The dataset is divided into training and testing sets.

- Training Set – Used to train the model.
- Testing Set – Used to evaluate model performance.

```python
# Split the dataset into training data and testing data
# Testing data is used to check how accurate the model is

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=45)
```

# Building the Logistic Regression Model

A Logistic Regression model is created to classify passengers as survived or not survived.

The `max_iter` parameter specifies the maximum number of iterations allowed during model training.

```python
# max_iter means the maximum number of times the model will try to learn from the data
# Increasing it helps the model finish training without errors

model = LogisticRegression(max_iter=1000)
```

# Model Training

The Logistic Regression model learns patterns from the training dataset to predict passenger survival.

```python
# Train the model using the training data
# The model learns the relationship between passenger details (X) and survival (Y)

model.fit(x_train, y_train)
```

# Model Prediction

The trained model predicts survival outcomes for passengers in the testing dataset.

```python
# Use the trained model to predict survival for test data

predictions = model.predict(x_test)
```

# Model Evaluation

The performance of the Logistic Regression model is evaluated using **Accuracy Score**, which measures how many predictions are correct.

```python
# Check how accurate the model predictions are

accuracy = accuracy_score(y_test, predictions)

print(" Highest Accuracy:", accuracy)
```
