Topics include:

These operations are commonly used during data preprocessing.

````python
#Indexing and Slicing in numpy
````

````python
ar= np.arange(1,11) #ar means array
````

````python
ar
````

````python
ar[8]  #value visualize with int 64 bytes / if float the float
````

````python
ar[-7] #Negative Indexing
````

````python
ar
````

````python
ar[1:8:2] #Slicing [Start:End:Step(take every 2nd element)]
````

````python
ar
````

````python
ar[1:10:2]
````

````python
ar[0:7:-1] #You are asking Python to go backward from index 0 to index 7, which is impossible.So Python returns an empty array
````

````python
ar[-1:3:-1] # start from last element, move backward, stop before index 3
````
